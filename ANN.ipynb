#Read the data in CSV file

```python
import pandas as pd

df = pd.read_csv("Mobile_Price_Classification-220531-204702.csv" , encoding = "latin-1")
df.head(5)
```

```python
df.shape
```

# Perprocessing

```python
 import nltk
 from nltk.stem.porter import PorterStemmer

 nltk.download('stopwords')
 from nltk.corpus import stopwords
 nltk.download('punkt')
 ps = PorterStemmer()
```

```python
df = df.drop_duplicates(keep = 'first')
```

#Split the data set into training and testing data where it contains 75% of the data as training and rest as testing data

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x = df.iloc[:, :-1]
y = df.iloc[:, -1]

x_train , x_test , y_train , y_test = train_test_split( x , y , test_size = 0.75 , random_state = 2)
```

```python
x_train.shape
```

# Ann model architecture

```python
import tensorflow as tf
from tensorflow import keras
```

```python
ann = tf.keras.models.Sequential()
```

# hidden layers

```python
ann.add(tf.keras.layers.Dense(units = 8 , activation = 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units = 4, activation = 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(units = 1 , activation = 'sigmoid'))
```

# compiling the ANN

```python
ann.compile(optimizer = 'adam' , loss = 'binary_crossentropy' ,metrics = ['accuracy'])
```

# train the model

```python
ann.fit(x_train,y_train, validation_data = (x_test,y_test), epochs = 100, batch_size = 32)
```

```python
from sklearn.metrics import accuracy_score

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(accuracy_score(y_test,y_pred))


from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()
model1.fit(x_train,y_train)
y_pred = model1.predict(x_test)
print(accuracy_score(y_test,y_pred))
```
